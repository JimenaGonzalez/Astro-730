```python
import numpy as np
import scipy.integrate
from matplotlib import pyplot as plt
```

```python
def E_parameter(a):
    return (np.sqrt(
        ome_rad*a**-4 + ome_m*a**-3 + ome_k*a**-2 
        + ome_DE*a**(-3*(1+ome_0 + ome_a))*np.exp(-3*ome_a*(1-a))))
```

```python
#Cosmology
ome_rad=0
ome_m=0.23
ome_k=0
ome_DE=0.73
ome_0=1 
ome_a=1 
```

```python
def E_parameter(a):
    return (np.sqrt(
        ome_rad*a**-4 + ome_m*a**-3 + ome_k*a**-2 + ome_DE))
```

```python
z = np.linspace(0,3,1000)
a = 1/(1+z)
param = E_parameter(a)
plt.plot(z,param)
plt.xlabel("z")
plt.ylabel(r'$E(z)=\frac{H(z)}{H_0}$')
plt.show()
```

```python
def Expansion_rate(a):
    return (np.sqrt(
        ome_rad*a**-4 + ome_m*a**-3 + ome_k*a**-2 + ome_DE)*a)
```

```python
z = np.linspace(0,3,1000)
a = 1/(1+z)
param = Expansion_rate(a)
plt.plot(z,param)
plt.xlabel("z")
plt.ylabel(r'Expansion rate = $H(z)a=\dot{a}$')
plt.show()
```
